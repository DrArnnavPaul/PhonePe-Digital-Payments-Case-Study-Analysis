# Task 3: Data Quality Checks
## 3.1: Ensure data consistency across state and district levels

```python

##1. For each state, calculate the total number of transactions, total transaction amount, and total registered users by summing up the values from the district level data. 
district_summary = (Distxn.groupby("State").agg(total_transactions=("Transactions", "sum"),total_transaction_amount=("Amount (INR)", "sum"),total_registered_users=("Registered Users", "sum")).reset_index())
print(district_summary)
```

```python
##2. Compare the results with the corresponding values at the state level to ensure they match.
state_summary = (statetx.groupby("State").agg(state_transactions=("Transactions", "sum"),state_transaction_amount=("Amount (INR)", "sum"),state_registered_users=("Registered Users", "sum")).reset_index())
comparison = pd.merge(district_summary,state_summary,on="State",how="inner")
comparison["transactions_match"] = (comparison["total_transactions"] == comparison["state_transactions"])

comparison["amount_match"] = (comparison["total_transaction_amount"] == comparison["state_transaction_amount"])

comparison["users_match"] = (comparison["total_registered_users"] == comparison["state_registered_users"])

print(comparison)
```

```python
##3. Display any discrepancies found between the district-level and state-level data.
discrepancies = comparison[(comparison["transactions_match"] == False) |(comparison["amount_match"] == False) |(comparison["users_match"] == False)]

print(discrepancies)
```

# Task 4: Data Merging and Advanced Analysis 
## 4.1: Ratio of users to population by state 

```python

##1. Merge the State_Txn and Users dataset with the District Demographics dataset to calculate the ratio of registered users to the population for each state. Display the results in a tabular format.
state_group = (statetx.groupby(['State'])['Registered Users'].sum().reset_index())
demo_group = (Disdemo.groupby(['State'])['Population'].sum().reset_index())
Ratio = state_group.merge(demo_group,on=['State'])
Ratio['Ratio'] = (Ratio['Registered Users'] /Ratio['Population'])
print(Ratio)
```

```python
# 2 - Create a column chart depicting the ratio of users to population by state.

plt.figure(figsize=(15, 4))
sns.barplot(data=Ratio.sort_values('Ratio', ascending=False),x='State',y='Ratio')

plt.xticks(rotation=90)

plt.title('Ratio of Users to Population by State')

plt.show()
```

## 4.2: Correlate population density with transaction volume 

```python

##1. Merge the District_Txn and Users dataset with the District Demographics dataset.
pt = Distxn.merge(Disdemo,on=['State', 'District', 'Code'])
dt = pt.groupby(['State', 'District']).agg({'Transactions': 'sum','Density': 'max'}).reset_index()
print(dt)
```

```python
##2: Correlate population density with transaction volume
correlation = dt['Density'].corr(dt['Transactions'])

print("Correlation between Population Density and Transaction Volume:")
print(correlation)
```

```python

## 3 - Create a scatter plot to visualize the correlation between population density and transaction volume.
plt.figure(figsize=(14,8))

sns.scatterplot(
    data=dt,
    x='Density',
    y='Transactions',
    hue='State')
plt.title('Population Density vs Transaction Volume')
plt.legend(
    bbox_to_anchor=(1.05, 1),
    loc='upper left'
)

plt.tight_layout()

plt.show()
```

```python
##4.3: Average transaction amount per user 
##1. Merge relevant datasets to calculate the average transaction amount per user for each state. Display the results in a tabular format.
avg_txn_user = statetx.groupby('State').agg({'Amount (INR)': 'sum','Registered Users': 'sum'}).reset_index()
avg_txn_user['avg_transaction_per_user'] = (
    avg_txn_user['Amount (INR)'] /
    avg_txn_user['Registered Users']
)
print(avg_txn_user)
```

```python
##2. Identify the top 5 states with the highest average transaction amount per user 
##    and the top 5 states with the lowest average transaction amount per user. Display the results.
top5 = avg_txn_user.nlargest(5,'avg_transaction_per_user')
print("\nTop 5 States with Highest Average Transaction Amount Per User:\n")
print(top5)
```

```python
##    and the top 5 states with the lowest average transaction amount per
low5 = avg_txn_user.nsmallest(5,'avg_transaction_per_user')
print("\nTop 5 States with Lowest Average Transaction Amount Per User:\n")
print(low5)
```

## 4.4: Device brand usage ratio


```python

##1. Merge the State_DeviceData dataset with the State_Txn and Users dataset.
device_usage = Devicedata.merge(
    statetx,
    on=['State', 'Year', 'Quarter']
)
device_usage
```

```python
##2. Calculate the ratio of users using each device brand to the total number of registered users in each state. Display the results in a tabular format.

##3. Create a bar chart depicting the device brand usage ratio for each state.

brand_users = device_usage.groupby(['State', 'Brand']).agg({'Registered Users_x': 'sum','Registered Users_y': 'sum'}).reset_index()

brand_users.rename(columns={'Registered Users_x': 'brand_users','Registered Users_y': 'total_registered_users'}, inplace=True)

brand_users['device_usage_ratio'] = (brand_users['brand_users'] /brand_users['total_registered_users'])

plt.figure(figsize=(15,6))

sns.barplot(
    data=brand_users,
    x='State',
    y='device_usage_ratio',
    hue='Brand'
)

plt.title('Device Brand Usage Ratio by State')

plt.xticks(rotation=90)

plt.legend(
    bbox_to_anchor=(1.02, 1),
    loc='upper left'
)

plt.tight_layout()

plt.show()
```

## 6.2: Correlate demographic data with transaction data

```python

##1. Find correlations between demographic data and transaction data.
demo_txn = Distxn.merge(Disdemo,on=['State', 'District', 'Code'])
district_corr = demo_txn.groupby(['State', 'District']).agg({'Transactions': 'sum','Amount (INR)': 'sum','Density': 'max','Population': 'max'}).reset_index()
correlation_matrix = district_corr[['Transactions','Amount (INR)','Density','Population']].corr()

print(correlation_matrix)

plt.figure(figsize=(8,5))

sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')

plt.title('Correlation Between Demographic and Transaction Data')

plt.show()
```
